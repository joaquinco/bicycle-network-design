```python
import os

import bcnetwork as bc
```

# Load graph

```python
base_path = '../instances/simple'
nodes_csv = os.path.join(base_path, 'nodes.csv')
arcs_csv = os.path.join(base_path, 'arcs.csv')
```

```python
g = bc.persistance.read_graph_from_csvs(nodes_csv, arcs_csv)

print('#nodes: ', g.number_of_nodes())
print('#arcs: ', g.number_of_edges())
```

```python
bc.draw.draw_graph(g)
```

# Problem setup

```python
budget = 50

model = bc.model.RandomModel(
    graph=g,
    budget=budget,
    odpair_count=8,
    breakpoint_count=3,
    project_root='..',
)
```

## Solve

```python
solution = model.solve()
```
